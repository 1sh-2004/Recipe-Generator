```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MultiLabelBinarizer

# Step 1: Load the training dataset
train_data = pd.read_csv("book1csv.csv")  

# Convert string lists into actual Python lists (if necessary)
train_data['ingredients_list'] = train_data['ingredients_list'].apply(lambda x: eval(x) if isinstance(x, str) else x)

# Step 2: Use MultiLabelBinarizer to One-Hot Encode Ingredients
mlb = MultiLabelBinarizer()
ingredients_encoded = mlb.fit_transform(train_data['ingredients_list'])
ingredients_df = pd.DataFrame(ingredients_encoded, columns=mlb.classes_)

# Combine with the target column
X = ingredients_df  # Features
y = train_data['Taste']  # Target

# Encode target labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Train-test split (optional)
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Step 3: Train the Model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Step 4: Load the new dataset
new_data = pd.read_csv("recipe_final (1).csv")  

# Convert new dataset ingredients column to lists
new_data['ingredients_list'] = new_data['ingredients_list'].apply(lambda x: eval(x) if isinstance(x, str) else x)

# Apply the same one-hot encoding transformation
new_ingredients_encoded = mlb.transform(new_data['ingredients_list'])
new_ingredients_df = pd.DataFrame(new_ingredients_encoded, columns=mlb.classes_)

# Step 5: Predict taste for the new dataset
predicted_taste = model.predict(new_ingredients_df)

# Convert predictions back to labels
new_data["predicted_taste"] = label_encoder.inverse_transform(predicted_taste)

# Step 6: Save the new dataset with predictions
new_data.to_csv("new_dataset_with_taste.csv", index=False)

print("Predictions added and saved successfully!")
```

```python
dff=pd.read_csv('new_dataset_with_taste.csv')
dff
```

```python
vectorizer = TfidfVectorizer()
X_ingredients = vectorizer.fit_transform(dff['ingredients_list'])
X_ingredients
```

```python
scaler = StandardScaler()
X_numerical = scaler.fit_transform(dff[['calories', 'fat', 'carbohydrates', 'protein', 'cholesterol', 'sodium', 'fiber',]])
X_numerical
```

```python
# Combine Features
X_combined = np.hstack([X_numerical, X_ingredients.toarray()])

# Train KNN Model
knn = NearestNeighbors(n_neighbors=3, metric='euclidean')
knn.fit(X_combined)
```

```python
def recommend_recipes(input_features):
    input_features_scaled = scaler.transform([input_features[:7]])
    input_ingredients_transformed = vectorizer.transform([input_features[7]])
    input_combined = np.hstack([input_features_scaled, input_ingredients_transformed.toarray()])
    distances, indices = knn.kneighbors(input_combined)
    recommendations = dff.iloc[indices[0]]
    result_df = dff[dff['recipe_name'].isin(recommendations['recipe_name'])]
    return result_df[['recipe_name', 'ingredients_list','calories','fat','protein','carbohydrates', 'image_url','predicted_taste']]

# Example Input
input_features = [15, 36, 1, 42, 21, 81, 2, 'pork belly']
recommendations = recommend_recipes(input_features)
recommendations
# result_df = dff[dff['recipe_name'].isin(recommendations['recipe_name'])]
# result_df
```
